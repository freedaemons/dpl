```python
import pandas as pd
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = 'NonStdAdd.tsv'
```

```python
df=pd.read_table(file)
```

```python
df.tail()
```

```python
cm = df['Count'].unique()
cm.sort()
cm.tolist()
```

```python
cmdict = {key: 0 for key in cm}
from pprint import pprint
pprint(cmdict)
```

```python
for row in df.itertuples():
    cmdict[row.Count] += 1
pprint(cmdict)
```

```python
print(len(cmdict))
print(sum(cmdict.values()))
cmdf = pd.DataFrame([[key,value] for key,value in cmdict.items()],columns=['plotcount','num'])
cmdf.sort_values(['plotcount','num'],ascending=[1,0], inplace=True)
cmdf.reset_index(drop=True,inplace=True)
pprint(cmdf)
```

```python
cmdf[['plotcount','num']].plot(kind='bar',figsize=(15,10))
plt.show()
```

```python
cmdfplot = cmdf.plot(x='plotcount',y='num')
```
